## Making figures with NumPy arrays

we want :
* Line plot
* Scatter plot
* Bar Plot
* Histogram
* subplots

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data 
fig, ax = plt.subplots()
ax.plot(x, np.sin(x));
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# make plot form dictionary
nut_butter_prices = {
    'Almond Butter': 10,
    'Peanut Butter': 8,
    'Cashew Butter': 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values())
```

```python
# make some data 
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# 2 Options for subplots

```python
# subplot options 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# plot to each different axis
ax1.plot(x, np.sin(x))
ax2.scatter(np.random.rand(1000), np.sin(x))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```
