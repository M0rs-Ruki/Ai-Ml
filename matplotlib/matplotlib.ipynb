```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
```

```python
plt.plot(x);
```

```python
plt.plot(y)
```

```python
plt.plot(x,y);
```

```python
# 1st Method
fig = plt.figure() # create figure
ax = fig.add_subplot() # add axes
plt.show()
```

```python
# 2nd Method
fig = plt.figure() # create figure
ax = fig.add_axes([1,1,1,1]) # left, bottom, width, height
ax.plot(x,y);
```

```python
# 3rd Method (Recommended)
fig, ax = plt.subplots() # create figure and axes
ax.plot(x,y);
```
