# 2 Options for subplots

```python
# subplot options 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# plot to each different axis
ax1.plot(x, np.sin(x))
ax2.scatter(np.random.rand(1000), np.sin(x))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
# plot to each different axis (index)
ax[0,0].plot(x, np.sin(x))
ax[0,1].scatter(np.random.rand(1000), np.sin(x))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

# Which one should use (Pyplot vs Object-Oriented)

* When plotting something quickly, use the Pyplot method.
* When plotting something more advanced, use the Object-Oriented method.


```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease.age > 50 ]
over_50.head()
```

```python
len(over_50)
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target'
             );
```

```python
# OO Method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax
             );

ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(over_50["age"], over_50["chol"], c=over_50["target"])
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol Level")
# Add a legend
ax.legend(*ax.collections[0].legend_elements(), title="Target");
# Add a horizontal line
middle = over_50["chol"].median()
ax.axhline(y=middle, color='r', linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)
                    
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol Level")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(), color='r', linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                        y=over_50["thalach"],
                        c=over_50["target"])
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(), color='r', linestyle='--');
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
